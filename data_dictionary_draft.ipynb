```python
import pandas as pd
from textwrap import shorten
df = pd.read_csv("data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
def guess_dtype(series: pd.Series) -> str:
    """Return tidy dtype label for the data dictionary."""
    if pd.api.types.is_bool_dtype(series):
        return "bool"
    if pd.api.types.is_integer_dtype(series):
        # 仅含 0/1 视作二元信号
        return "int (0/1)" if set(series.dropna().unique()) <= {0, 1} else "int"
    if pd.api.types.is_float_dtype(series):
        return "float"
    return "string"  # 默认

def sample_val(series: pd.Series) -> str:
    val = series.dropna().iloc[0] if series.dropna().size else ""
    return shorten(str(val), width=12, placeholder="…")

# 初步信息表
info_df = pd.DataFrame({
    "Field Name": df.columns,
    "Data Type": [guess_dtype(df[c]) for c in df.columns],
    "Missing Count": df.isna().sum(),
    "Example Value": [sample_val(df[c]) for c in df.columns]
})

# 识别所有二元或有限级别的字符串列，改标为 category
cat_candidates = [
    c for c in df.columns
    if df[c].dtype == "object" and df[c].nunique() <= 10   # 阈值可调
]
info_df.loc[info_df["Field Name"].isin(cat_candidates), "Data Type"] = "category"

# 保存草稿
info_df[["Field Name", "Data Type", "Example Value"]].to_markdown(
    "data_dictionary_draft.md", index=False
)
print("✅  data_dictionary_draft.md generated")
```
